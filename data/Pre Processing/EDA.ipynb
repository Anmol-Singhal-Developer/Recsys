```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
restaurants = pd.read_excel('Data_Restaurants.xlsx')
recipes     = pd.read_excel('Final_Recipe_Cleaned.xlsx')
```

```python
def missing_summary(df):
    ms  = df.isnull().sum()
    pct = (ms / len(df) * 100).round(2)
    return pd.DataFrame({
        'missing_count': ms,
        'missing_pct':   pct
    }).sort_values('missing_pct', ascending=False)

print("Restaurants Missing")
print(missing_summary(restaurants), "\n")
print("Recipes Missing")
print(missing_summary(recipes), "\n")
```

```python
print("Restaurants Numeric Summary")
print(restaurants.select_dtypes(include='number').describe().T, "\n")

print("Recipes Numeric Summary")
print(recipes[['Recipe Servings', 'Rating']].describe().T, "\n")
```

```python
plt.figure()
restaurants['AggregateRating'].dropna().hist(bins=30)
plt.title('Restaurant Ratings')
plt.xlabel('Rating'); plt.ylabel('Count')
plt.show()

plt.figure()
recipes['Rating'].dropna().hist(bins=30)
plt.title('Recipe Ratings')
plt.xlabel('Rating'); plt.ylabel('Count')
plt.show()

plt.figure()
recipes['Recipe Servings'].dropna().hist(bins=30)
plt.title('Recipe Servings')
plt.xlabel('Servings'); plt.ylabel('Count')
plt.show()
```
